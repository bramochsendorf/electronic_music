```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
plt.close('all')
```

```python
# genres and colors to plot    
genre = ['electro','house','techno']
color = ['red','blue','orange']
year_max = 2016

# binning in years (optional).
bin_years = 1
```

```python
# plot the results!    
xaxis = np.arange(1980,np.int_(year_max)+1,bin_years)

# counter 
i = 0 
        
fig = plt.figure(figsize=(10.5,12))
font = {'size':13}
plt.rc('font', **font)
plt.subplot(321)

plt.subplots_adjust(left=0.2, bottom=0.2, right=0.875, top=0.85, wspace=0.35, hspace=0.35)

for gen in genre:
    
    result = pd.read_pickle('data/'+gen+'_stats')
    
    # this is for when we'd like to bin data over years to reduce noise
    result2 = ((result + result.shift(-1)) / 2)[::bin_years]
    
    # not plot errorbars else the plot overpopulated 
    plt.subplot(321)
    plt.step(xaxis,result2['popu'],color=color[i])
    plt.xlabel('Release year')
    plt.xlim([1980,2016])
    plt.ylabel('Popularity')
    plt.legend(genre)
    
    plt.subplot(322)
    plt.step(xaxis,result2['temp'],color=color[i])
    plt.xlabel('Release year')
    plt.xlim([1980,2016])
    plt.ylabel('Beats per minute')

    plt.subplot(323)
    plt.step(xaxis,result2['loud'],color=color[i])
    plt.xlabel('Release year')
    plt.xlim([1980,2016])
    plt.ylabel('Loudness (dB)')
    
    plt.subplot(324)
    plt.step(xaxis,result2['dance'],color=color[i])
    plt.xlabel('Release year')
    plt.xlim([1980,2016])
    plt.ylabel('Danceability')
    
    plt.subplot(325)
    plt.step(xaxis,result2['inst'],color=color[i])
    plt.xlabel('Release year')
    plt.xlim([1980,2016])
    plt.ylabel('Instrumentalness')
    
    plt.subplot(326)
    plt.step(xaxis,result2['val'],color=color[i])
    plt.xlabel('Release year')
    plt.xlim([1980,2016])
    plt.ylabel('Valence')
    
    i += 1

plt.show()
#plt.savefig('output/combined_evolution.pdf', dpi=150, bbox_inches='tight', pad_inches=0.2)
```
